## Četvorougao sa najsličnijom površinom



```python
import matplotlib.pyplot as plt  
import random
%matplotlib inline

# diff((p2 - p0), (p1-p0))

def isCC(p0, p1, p2):
    diff = ((p2[0]-p0[0])*(p1[1]-p0[1])-
            (p1[0]-p0[0])*(p2[1]-p0[1]))
    return diff>0



points = []
for _ in range(50):
    [x, y] = [random.randint(-100, 100), random.randint(-100, 100)]
    points.append([x, y])
    
```

```python
x = [p[0] for p in points]
y = [p[1] for p in points]
plt.plot(x, y, marker='D', linestyle='None')

plt.xlim([-150, 150])
plt.ylim([-150, 150])
```

```python
a = random.choice(points)
b = random.choice(points)
```

```python
hx = [a[0], b[0]]
hy = [a[1], b[1]]

#filter levi

levi = []
for p in points:
    if isCC(a, b, p):
        levi.append(p)

```

```python
x = [p[0] for p in points]
y = [p[1] for p in points]
plt.plot(x, y, marker='D', linestyle='None')

plt.xlim([-150, 150])
plt.ylim([-150, 150])

plt.plot(hx, hy)

xx = [p[0] for p in levi]
yy = [p[1] for p in levi]
plt.plot(xx, yy, 'ro')

```
