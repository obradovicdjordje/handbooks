```python
import matplotlib.pyplot as plt  
import random
%matplotlib inline
```

```python
points = []
for _ in range(3):
    [x, y] = [random.randint(-100, 100), random.randint(-100, 100)]
    points.append([x, y])
    
x = [p[0] for p in points]
y = [p[1] for p in points]
x.append(points[0][0])
y.append(points[0][1])
plt.plot(x, y, marker='D')
plt.grid(True)

m = [points[0][0], points[0][1]]
M = [points[0][0], points[0][1]]
for p in points:
    for j in [0,1]:
        if p[j]<m[j]:
            m[j] = p[j]
        if p[j]>M[j]:
            M[j] = p[j]

A = [m[0], m[1]]
B = [m[0], M[1]]
C = [M[0], M[1]]
D = [M[0], m[1]]
            
bbox = [A, B, C, D, A]
            
hx = [p[0] for p in bbox]
hy = [p[1] for p in bbox]
plt.plot(hx, hy)

plt.xlim([m[0]-5, M[0]+5])
plt.ylim([m[1]-5, M[1]+5])
```
